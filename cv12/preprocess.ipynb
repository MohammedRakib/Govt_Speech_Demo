```python
import pandas as pd
import os

TRAIN_PATH = os.path.join(os.path.abspath("."), "train.tsv")
VAL_PATH = os.path.join(os.path.abspath("."), "dev.tsv")
TEST_PATH = os.path.join(os.path.abspath("."), "test.tsv")
OTHER_PATH = os.path.join(os.path.abspath("."), "other.tsv")
```

```python
df = pd.read_csv(TRAIN_PATH, sep="\t")
df2 = pd.read_csv(VAL_PATH, sep="\t")
df3 = pd.read_csv(TEST_PATH, sep="\t")
df.head()
```

```python
df.drop(['client_id'], axis=1, inplace=True)
df2.drop(['client_id'], axis=1, inplace=True)
df3.drop(['client_id'], axis=1, inplace=True)
df.head()
```

```python
df['path'] = df['path'].apply(lambda x: os.path.join("clips", x))
df2['path'] = df2['path'].apply(lambda x: os.path.join("clips", x))
df3['path'] = df3['path'].apply(lambda x: os.path.join("clips", x))
df.head()
```

```python
df.rename(columns = {'path':'file_name'}, inplace = True)
df2.rename(columns = {'path':'file_name'}, inplace = True)
df3.rename(columns = {'path':'file_name'}, inplace = True)
```
